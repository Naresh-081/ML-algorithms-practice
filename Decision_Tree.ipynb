```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.DataFrame(load_breast_cancer()['data'],columns=load_breast_cancer()['feature_names'])
df['target']=load_breast_cancer()['target']
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('target',axis=1),df['target'],test_size=0.2, random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier
cls = DecisionTreeClassifier()
cls.fit(X_train,y_train)
```

```python
predcition= cls.predict(X_test)
predcition
```

```python
cls.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,precision_score
print(classification_report(y_test,predcition))
print(confusion_matrix(y_test,predcition))
print(accuracy_score(y_test,predcition))
print(precision_score(y_test,predcition))
```

```python
feature_names = df.drop('target',axis=1).columns.tolist()
feature_names
```

```python
cls.feature_importances_
```

```python
importance=pd.DataFrame(cls.feature_importances_,index=feature_names,columns=['importance'])
importance.sort_values('importance',ascending=False)
```

```python
importance.plot(kind='bar')
plt.show()
```

```python
from sklearn import tree
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25, 20))
_ = tree.plot_tree(cls,
                   feature_names=feature_names,
                   class_names={0: 'Malignant', 1: 'Benign'},
                   filled=True,
                   fontsize=12)
```
